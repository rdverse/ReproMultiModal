```
def build_cnn_model40(LR=0.1,Frame_size=150):
    #model = keras.Sequential()
    
    input1 = layers.Input(shape = (3,Frame_size,1))
    x1 = layers.Conv2D(filters = 9,kernel_size=(3,3),padding='same', activation='relu')(input1)
    x11 = layers.GlobalMaxPool2D()(x1)
    
    input2 = layers.Input(shape = ( 3,Frame_size,1))
    x2 = layers.Conv2D(filters = 9,kernel_size=(3,3),padding='same', activation='relu')(input2)
    x22 = layers.GlobalMaxPool2D()(x2)

    merged = layers.Concatenate()([x11,x22])
    
    next_one = layers.Dense(150, activation='relu')(merged)
    next_two = layers.Dropout(0.5)(next_one)
    next_three = layers.Dense(30, activation='relu')(next_two)
    next_four = layers.Dropout(0.3)(next_three)
    next_five = layers.Dense(1)(next_four)


    # from keras.models import Model
    model = tf.keras.Model(inputs = [input1,input2], outputs = next_five)

    optimizer = tf.keras.optimizers.RMSprop(LR)
    model.compile(optimizer = optimizer,
              loss='mean_absolute_error',
              metrics=['mae'])
    return model

```

```
def build_cnn_model4(LR=0.1,Frame_size=150):
    
    input1 = layers.Input(shape = ( 3,Frame_size,1))
    x1 = layers.Conv2D(filters = 9,kernel_size=(3,3),padding='same', activation='relu')(input1)
    x11 = layers.GlobalMaxPool2D()(x1)
    

    input2 = layers.Input(shape = ( 3,Frame_size,1))
    x2 = layers.Conv2D(filters = 9,kernel_size=(3,3),padding='same', activation='relu')(input2)
    x22 = layers.GlobalMaxPool2D()(x2)
    
    merged = layers.Concatenate()([x11,x22])
    
    
    next_one = layers.Dense(150, activation='relu')(merged)
    next_two = layers.Dropout(0.5)(next_one)
    next_three = layers.Dense(30, activation='relu')(next_two)
    next_four = layers.Dropout(0.3)(next_three)
    next_five = layers.Dense(1)(next_four)
    
    model = tf.keras.Model(inputs = [input1,input2], outputs = next_five)

    optimizer = tf.keras.optimizers.RMSprop(LR)
    model.compile(optimizer = optimizer,
              loss='mean_absolute_error',
              metrics=['mae'])
    return model
```
